**Feature Engineering**

```python

diabetes_data = pd.read_csv('diabetes.csv')

# Create categorical BMI ranges
bmi_bins = [0, 18.5, 24.9, 29.9, float('inf')]
bmi_labels = ['underweight', 'normal weight', 'overweight', 'obese']
diabetes_data['bmi_category'] = pd.cut(diabetes_data['BMI'], bins=bmi_bins, labels=bmi_labels)

# Create age groups
age_bins = [0, 30, 60, float('inf')]
age_labels = ['young', 'middle-aged', 'senior']
diabetes_data['age_group'] = pd.cut(diabetes_data['Age'], bins=age_bins, labels=age_labels)

# Create glucose-insulin interaction feature
diabetes_data['glucose_insulin_interaction'] = diabetes_data['Glucose'] * diabetes_data['Insulin']

# Create blood pressure categories
bp_bins = [0, 80, 90, 120, float('inf')]
bp_labels = ['low', 'normal', 'elevated', 'high']
diabetes_data['bp_category'] = pd.cut(diabetes_data['BloodPressure'], bins=bp_bins, labels=bp_labels)

# Create an interaction between pregnancies and insulin
diabetes_data['pregnancies_insulin_interaction'] = diabetes_data['Pregnancies'] * diabetes_data['Insulin']

# Calculate age-BMI interaction
diabetes_data['age_bmi_interaction'] = diabetes_data['Age'] * diabetes_data['BMI']

# Calculate insulin-glucose ratio
diabetes_data['insulin_glucose_ratio'] = diabetes_data['Insulin'] / diabetes_data['Glucose']

# Display the modified dataset
diabetes_data.head()
```

# 3.Model Building
**Splitting the dataset**

```python
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']
```

**Now we will split the data into training and testing data using the train_test_split function**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,
                                                    random_state=7)
```

# Random Forest
**Building the model using RandomForest**

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, y_train)
```

**Now after building the model let’s check the accuracy of the model on the training dataset.**

```python
rfc_train = rfc.predict(X_train)
from sklearn import metrics

print("Accuracy_Score =", format(metrics.accuracy_score(y_train, rfc_train)))
```

**So here we can see that on the training dataset our model is overfitted.**

**Getting the accuracy score for Random Forest**

```python
from sklearn import metrics

predictions = rfc.predict(X_test)
print("Accuracy_Score =", format(metrics.accuracy_score(y_test, predictions)))
```

**Classification report and confusion matrix of random forest model**

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test,predictions))
```

# Decision Tree
**Building the model using DecisionTree**

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

Now we will be making the predictions on the testing data directly as it is of more importance.

**Getting the accuracy score for Decision Tree**

```python
from sklearn import metrics

predictions = dtree.predict(X_test)
print("Accuracy Score =", format(metrics.accuracy_score(y_test,predictions)))
```

**Classification report and confusion matrix of the decision tree model**

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test,predictions))
```

# Support Vector Machine (SVM)
**Building the model using Support Vector Machine (SVM)**

```python
from sklearn.svm import SVC

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

**Prediction from support vector machine model on the testing data**

```python
svc_pred = svc_model.predict(X_test)
```

**Accuracy score for SVM**

```python
print("Accuracy Score =", format(metrics.accuracy_score(y_test, svc_pred)))
```

**Classification report and confusion matrix of the SVM classifier**

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, svc_pred))
print(classification_report(y_test,svc_pred))
```
